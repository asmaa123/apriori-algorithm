```python
df['Description'] = df['Description'].str.strip()
df['InvoiceNo'] = df['InvoiceNo'].astype('str')  
df1=df[~df['InvoiceNo'].str.contains('c')]
```

```python
countries=df['Country'].unique().tolist()
```

```python
countries
```

```python
for i,country in enumerate(countries):
             basket= df[df['Country'] ==countries[i]].groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo')
            #***********************
             def hot_encode1(P):  
                if(P<= 0):  
                    return 0  
                if(P>= 1):  
                    return 1
             basket1_encoded = basket.applymap(hot_encode1)  
             basket = basket1_encoded 
             frq_items1 = apriori(basket, min_support = 0.05, use_colnames = True)  
             rules1 = association_rules(frq_items1, metric ="lift", min_threshold = 1)  
             rules1 = rules1.sort_values(['confidence', 'lift'], ascending =[False, False]) 
             print('For :     '+countries[i])
             print(rules1.head())
             print("**********************************************************************************************************")
             time.sleep(60)
```
